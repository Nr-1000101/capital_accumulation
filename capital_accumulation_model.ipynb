```python
import matplotlib.pyplot as plt

# Recursive function definitions
def red_monthly_recursive(n, values, yearly_inflation_rate, yearly_investment_returns):
    if n == 0:
        return 0  # starting value
    return values[n-1] * (1 - yearly_inflation_rate ** 1/12) * (1 + yearly_investment_returns ** 1/12) + salary_red
                          
def blue_monthly_recursive(n, values, yearly_inflation_rate):
    if n == 0:
        return 0  # starting value
    return values[n-1] * (1 - yearly_inflation_rate ** 1/12) + salary_blue

def red_yearly_recursive(n, values, yearly_inflation_rate, yearly_investment_returns):
    if n == 0:
        return 0  # starting value
    return values[n-1] * (1 - yearly_inflation_rate) * (1 + yearly_investment_returns) + 12 * salary_red
                          
def blue_yearly_recursive(n, values, yearly_inflation_rate):
    if n == 0:
        return 0  # starting value
    return values[n-1] * (1 - yearly_inflation_rate) + 12 * salary_blue


def savings_with_monthly_investment(years=10, salaray_red=1500, salary_blue=2500, yearly_inflation_rate =.03, yearly_investment_returns =.1):
    # Number of time steps (months)
    months = years * 12
    t = list(range(months + 1))
    
    # Generate y-values using recursion
    y_red = []
    y_blue = []
    
    for i in range(len(t)):
        y_red.append(red_monthly_recursive(i, y_red, yearly_inflation_rate, yearly_investment_returns))
        y_blue.append(blue_monthly_recursive(i, y_blue, yearly_inflation_rate))
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))
    
    ax.plot([time / 12 for time in t], y_red, color='red', label='1500 monthly salary - '+ str(yearly_inflation_rate * 100) + ' % yearly inflation + '
            + str(yearly_investment_returns * 100) + ' % yearly investment returns')
    ax.plot([time / 12 for time in t], y_blue, color='blue', label='2500 monthly salary - '+ str(yearly_inflation_rate * 100) + ' % yearly inflation')
    
    # Axis labels
    ax.set_xlabel("time (years)")
    ax.set_ylabel("capital")  # left blank
    
    # Add legend (explanation inside the graph)
    ax.legend(loc="upper left")

    plt.title("capital accumulation with monthly investment")
    
    plt.show()

def savings_with_yearly_investment(years=10, salaray_red=1500, salary_blue=2500, yearly_inflation_rate =.03, yearly_investment_returns =.1):
    # Number of time steps (months)
    t = list(range(years + 1))
    
    # Generate y-values using recursion
    y_red = []
    y_blue = []
    
    for i in range(len(t)):
        y_red.append(red_yearly_recursive(i, y_red, yearly_inflation_rate, yearly_investment_returns))
        y_blue.append(blue_yearly_recursive(i, y_blue, yearly_inflation_rate))
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))
    
    ax.plot(t, y_red, color='red', label='1500 monthly salary - '+ str(yearly_inflation_rate * 100) + ' % yearly inflation + '
            + str(yearly_investment_returns * 100) + ' % yearly investment returns')
    ax.plot(t, y_blue, color='blue', label='2500 monthly salary - '+ str(yearly_inflation_rate * 100) + ' % yearly inflation')
    
    # Axis labels
    ax.set_xlabel("time (years)")
    ax.set_ylabel("capital")  # left blank
    
    # Add legend (explanation inside the graph)
    ax.legend(loc="upper left")

    plt.title("capital accumulation with yearly investment")
    
    plt.show()

if __name__ == "__main__":
    savings_with_monthly_investment(30, 1500, 2500, .03, .15)
    savings_with_yearly_investment(30, 1500, 2500, .03, .15)
```
